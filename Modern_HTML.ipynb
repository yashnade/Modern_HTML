Q2. What are HTML Entities? List out 5 commonly used HTML entities

HTML Entities

HTML Entities are special codes used to represent characters that have a specific meaning in HTML or that cannot be directly typed on a keyboard. They help display reserved characters correctly in a web browser without affecting the HTML structure.

Entities start with an ampersand (&) and end with a semicolon (;). For example, the less-than symbol (<) is represented as &lt; in HTML to avoid confusion with HTML tags.

Five Commonly Used HTML Entities:

    &lt; – Represents the less-than symbol (<)

    &gt; – Represents the greater-than symbol (>)

    &amp; – Represents the ampersand symbol (&)

    &quot; – Represents a double quote (")

    &apos; – Represents a single quote or apostrophe (')

Q3.Define accessibility in the context of web development. Discuss why it's essential to create accessiblewebsites and how it benefits different user group

Ans:

Accessibility in Web Development

Accessibility in web development refers to the practice of designing and building websites that can be easily accessed and used by all individuals, including those with disabilities such as visual, auditory, cognitive, or motor impairments.

It involves ensuring that content is perceivable, operable, understandable, and robust for all users, regardless of their abilities or the devices they use.

Importance of Creating Accessible Websites

    Inclusivity:
    Accessible websites allow people with disabilities to interact with content and services equally.

    Legal Compliance:
    Many countries have laws (like the ADA, Section 508, or WCAG) that require websites to be accessible.

    Better User Experience:
    Accessibility improves navigation, readability, and overall usability for all users.

    SEO Benefits:
    Search engines favor accessible websites, improving visibility and ranking.

    Wider Reach:
    Accessible design ensures more users, including elderly people and those with temporary limitations (e.g., a broken arm), can use the website effectively.

Q4.List any 3 ways which help us in improving the accessibility of HTM1

Ans:

Three Ways to Improve Accessibility in HTML

    Use Semantic HTML Tags: Semantic elements like <header>, <nav>, <main>, <article>, and <footer> provide meaningful structure to a web page. They help screen readers and assistive technologies understand the layout and content more accurately.

    Add Alt Text to Images: The alt attribute in the <img> tag provides a text description of images. This helps visually impaired users understand the content of images through screen readers.

    Ensure Keyboard Navigation: Design websites so that all interactive elements (like buttons, links, and form fields) can be accessed and operated using the keyboard alone. This benefits users who cannot use a mouse.

Q5. Create a web page that highlights the features of HTML5. Use appropriate semantic tags to structure the
content and showcase at least three key features of HTML5 with explanations>


Ans:


```python
<!DOCTYPE html>
<html lang="en">
<head>
    <meta charset="UTF-8">
    <meta name="viewport" content="width=device-width, initial-scale=1.0">
    <title>HTML5 Features</title>
</head>
<body>

<header>
    <h1>Key Features of HTML5</h1>
    <p>Learn about the new and improved features of HTML5.</p>
</header>

<section>
    <article>
        <h2>1. New Semantic Tags</h2>
        <p>HTML5 introduced new semantic elements such as <code>&lt;header&gt;</code>, <code>&lt;footer&gt;</code>, <code>&lt;nav&gt;</code>, and <code>&lt;article&gt;</code>. These tags help define the structure of a webpage, making it easier for search engines and assistive technologies to understand and interpret the content.</p>
    </article>

    <article>
        <h2>2. Audio and Video Support</h2>
        <p>HTML5 introduced the <code>&lt;audio&gt;</code> and <code>&lt;video&gt;</code> tags, allowing native embedding of multimedia content without the need for third-party plugins like Flash. This makes it easier to integrate audio and video into web pages and enhances cross-browser compatibility.</p>
        <p>Example of embedding a video:</p>
        <video controls width="320">
            <source src="https://www.w3schools.com/html/mov_bbb.mp4" type="video/mp4">
            Your browser does not support the video tag.
        </video>
    </article>

    <article>
        <h2>3. Local Storage</h2>
        <p>HTML5 introduced local storage capabilities through the <code>localStorage</code> and <code>sessionStorage</code> objects. These allow web applications to store data locally in the user's browser, providing better performance and offline support.</p>
    </article>
</section>

<footer>
    <p>&copy; 2025 HTML5 Features</p>
</footer>

</body>
</html>
```
